```python
#importowanie bilbiotek
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# wczytywanie danych
data = pd.read_csv('/Salary Data (2).csv')
print(data)
```

```python
#podstawowe obliczenia, aby zobaczyć strutkurę danych
educational_level_count = data["Education Level"].value_counts()
print(educational_level_count)
```

```python
#wykres
avg_salary = data.groupby("Education Level")['Salary'].mean()

avg_salary.plot(kind='bar')
plt.xlabel("Education Level")
plt.ylabel("Avg. Salary")
plt.title("Avg. Salary vs Education Level")
plt.xticks(rotation = 0)
plt.show()
```
